```python
# En este caso, se usa el try para que el programa pueda concluir sin errores
try:
  n = float(input("Introduce un numero: "))
  m = 4
  print("{}/{}={}".format(n,m,n/m))
except:
  print("Ha ocurrido un error, Introduce bien el numero")
```

```python
# Este es un ciclo para poder hacer que el programa no avance hasta tener un valor correcto
# Sin ambargo tiene un defecto, aunque se ejecute correctamente se ejecutara de nuevo
while(True):
  try:
    n = float(input("Introduce un numero: "))
    m = 4
    print("{}/{}={}".format(n,m,n/m))
  except:
    print("Ha ocurrido un error, Introduce bien el numero")
```

```python
# para quitar ese problema, se hace de la siguiente forma
# Uso de Break y Else
while(True):
  try:
    n = float(input("Introduce un numero: "))
    m = 4
    print("{}/{}={}".format(n,m,n/m))
    #break # con este break se rompe la iteracion si todo salio bien.
  except:
    print("Ha ocurrido un error, Introduce bien el numero")
  else: #tambien la ventaja del manejo de excepciones es que se puede agregar un else
    # Se puede ver de manera que si no existe una excepcion se ejecuta la funcion else
    print("Todo a funcionado correctamente")
    break
```

```python
# Uso del Finally
while(True):
  try:
    n = float(input("Introduce un numero: "))
    m = 4
    print("{}/{}={}".format(n,m,n/m))
    #break # con este break se rompe la iteracion si todo salio bien.
  except:
    print("Ha ocurrido un error, Introduce bien el numero")
  else: #tambien la ventaja del manejo de excepciones es que se puede agregar un else
    # Se puede ver de manera que si no existe una excepcion se ejecuta la funcion else
    print("Todo a funcionado correctamente")
    break
  finally: # Se ejecuta siempre despues de una iteracion
    print("Fin de la Iteracion")
```

Manejor de erres.
1. TRY= capturar algun error
2. EXCEPT: definir codigo excepcional
3. ELSE: para definir un codigo que se ejecutara si no ocurre ningun error
4. FINALLY: definir un codigo que se ejecutara, alla o no un error.
