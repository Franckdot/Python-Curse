```python
v = "otro texto"
n = 10
print("texto ", v ,"y un numero ", n)
```

```python
c = "Un texto {} y un numero {}".format(v,n)
c
```

```python
print(c)
```

```python
print( "Un texto {1} y un numero {0}".format(v,n) )
```

```python
print( "Un texto {texto} y un numero {numero}".format(numero = n, texto = v) )
```

```python
print( "Un texto {n} y un numero {v}".format(v = v,n = n) )
```

```python
print("{v},{v},{v},{v}".format(v=v))
```

```python
#Alineacion a la derecha
print("{:>30}".format("palabra"))
```

```python
#Alineacion a la izquierda, le da 30 espacion en blanco.
print("{:30}".format("palabra"))
```

```python
# Alinea al centro
print("{:^30}".format("palabra"))
```

```python
# Truncamiento a 3 caracteres
print("{:.3}".format("palabra"))
```

```python
# Truncamiento a 3 caracteres y alineamiento a la derecha con 30 caracteres
print("{:>30.3}".format("palabra"))
```
