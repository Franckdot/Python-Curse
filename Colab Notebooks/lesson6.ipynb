```python
# revisa y si cumple alguna, y si alguna se cumple se queda con esa y descarta las demas opciones
nota = float(input("Introduce una calificacion: "))
if nota >= 9:
    print("Sobresaliente")
elif nota >= 7:
  print("Notable")
elif nota >= 6:
  print("Bien")
elif nota >= 5:
  print("Suficiente")
else:
  print("Necesitas estudiar mas")
```
