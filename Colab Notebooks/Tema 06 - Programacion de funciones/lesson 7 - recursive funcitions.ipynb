```python
# La recursividad es una tecnica utilizada para que cuando una actividad sea dificil
# se divida en partes mas pequeÃ±as que se puedan resolver.
# funciones recursiva sin retorno y con retorno
# Una caracteristicas de las funciones recursivas, esque se llaman a si mismas
def cuenta_atras(num):
  num -= 1
  if num > 0:
    print(num)
    cuenta_atras(num)
  else:
    print("Boooooom!")
  print("Fin de la funcion: ", num)

cuenta_atras(5)
```

```python
def factorial(num):
  if num > 1:
    num = num * factorial(num -1)
  return num

factorial(5)
```

```python

def factorial(num):
  print("Valor inicial ->", num)
  if num > 1:
    num = num * factorial(num -1)
  print("valor final -> ", num)
  return num

factorial(5)
```
