```python
# Los argumentos que se mandan a la funcion se utilizan como una copia en sus propias variables.
# Las listas funcionan por referencia, en lugar de una copia, se estara usando el dato original.
# y si los modificamos, tambien se vera reflejado en el dato exterior a la funcion.
def doblar_valor(numero):
  numero *= 2

n = 10
doblar_valor(n)
n
```

```python
def doblar_valores(numeros):
  for i,n in enumerate(numeros):
    numeros[i] *= 2

ns = [10,50,100]
doblar_valores(ns)
ns
```

```python
def doblar_valor(numero):
  return numero * 2
n = 10
n = doblar_valor(n)
n
```

```python
# El siguiente codigo nos ayuda a enviar una sublista usando slicing, se nevia una copia de la lista original
def doblar_valor(numeros):
  for i,n in enumerate(numeros):
    numeros[i] *= 2
ns = [10,50,100]
doblar_valor(ns[:])
ns
```
