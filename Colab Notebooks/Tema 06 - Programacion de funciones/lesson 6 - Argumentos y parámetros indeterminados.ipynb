```python
def indeterminados_posicion(*args):
  for arg in args:
    print(arg)

indeterminados_posicion(5, "Hola", [1,2,3,4,5])
```

```python
def indeterminados_nombre(**kwargs):
  print(kwargs)
indeterminados_nombre(n=5, c="Hola", l=[1,2,3,4,5])
```

```python
def indeterminados_nombre(**kwargs):
  for kwarg in kwargs:
    print(kwarg)
indeterminados_nombre(n=5, c="Hola", l=[1,2,3,4,5])
```

```python
def indeterminados_nombre(**kwargs):
  for kwarg in kwargs:
    print(kwarg," ",kwargs[kwarg])
indeterminados_nombre(n=5, c="Hola", l=[1,2,3,4,5])
```

```python
def super_function(*args,**kwargs):
  t = 0
  for arg in args:
    t += arg
  print("Sumatorio indeterminado es ", t)
  for kwarg in kwargs:
    print(kwarg," ",kwargs[kwarg])
super_function(10,50,-1,1.56,nombre="Hector",edad=27)
```
