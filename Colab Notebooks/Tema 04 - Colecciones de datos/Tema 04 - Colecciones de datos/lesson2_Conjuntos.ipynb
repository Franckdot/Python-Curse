```python
# Transformaremos una lista en un conjunto para aprovechar la eliminacion de duplicados
#Primero generamos la lista
estoUnaLista = [1,2,3,3,2,1]
estoUnaLista
```

```python
#Hacemos el cambio de una lista a un conjunto
estoEsUnConjunto = set(estoUnaLista)
estoEsUnConjunto
```

```python
#Si nuestro interes es seguir trabajando con la lista, volvemos a hacer el cambio
#vuala, quedo una lista sin duplicados
estoUnaLista = list(estoEsUnConjunto)
estoUnaLista
```

```python
estoUnaLista = [1,2,3,3,2,1]
estoUnaLista = list( set(estoUnaLista))
estoUnaLista
```
