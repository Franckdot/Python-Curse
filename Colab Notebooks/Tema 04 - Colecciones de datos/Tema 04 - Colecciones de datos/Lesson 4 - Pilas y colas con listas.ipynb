```python
#Colas: FIFO
#firt in, first out (primero en entrar, primero en salir)
from collections import deque
cola = deque()
cola
```

```python
cola = deque(['Hector','Juan','Miguel'])
cola
```

```python
cola.append('Maria')
cola.append('Arnaldo')
cola
```

```python
cola.popleft()
```

```python
p = cola.popleft()
p
```
