```python
edades = {'Hector':27,'Juan':45,'Maria':34}
edades
```

```python
#modificacion
edades['Hector'] += 1
edades
```

```python
edades['Juan'] + edades['Maria']
```

```python
for edad in edades:
  print(edad)
```

```python
for clave in edades:
  print(edades[clave])
```

```python
for clave in edades:
  print(clave,edades[clave])
```

```python
# Metodo item, una clave y un valor para cada registro
for clave,valor in edades.items():
  print(clave,valor)
```
